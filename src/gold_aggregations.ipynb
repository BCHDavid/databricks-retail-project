```python
from pyspark.sql.functions import sum, count

df = spark.table("main.retail.orders_silver")

df_agg = df.agg(
    sum("total_amount").alias("total_revenue"),
    count("order_id").alias("total_orders")
)

df_agg.write.format("delta").mode("overwrite").saveAsTable(
    "main.retail.orders_gold"
)

display(df_agg)
```
